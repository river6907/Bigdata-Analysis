
(8) 2007년도 데이터만 추출

```python
r1 = df[df.year == 2007] 
display("2007년도 데이터만 추출", r1)
```


(9)데이터 일부 추출, 한국(Korea Rep.)에 대한 2002,2007년 데이터만 추출 

```python
r2_korea = df[(df.country == 'Korea Rep.')]
r2 =  r2_korea[(r2_korea.year == 2002) | (r2_korea.year == 2007) ]
display("한국(Korea Rep.)에 대한 2002,2007년 데이터만 추출", r2)
```

(11) 년도순으로 데이터 정렬

```python
dfyear = df.sort_values(by='year') 
display(dfyear)
```

(12) 인구수 역순으로 데이터 정렬


```python
dfpop = df.sort_values(by='pop',ascending=False )
display(dfpop)
```

(13) 년도, 대륙순으로 데이터 정렬

```python
dfyear_conti = df.sort_values(['year','continent'])
dfyear_conti
```

(14) 대륙별로 그룹처리하여 그룹에 대한 중앙값, 평균, 표준편차 요약통계

```python
med = df.groupby('continent').median()
display("그룹에 대한 중앙값 =", med)
mean = df.groupby('continent').mean()
display("그룹에 대한 평균 =", mean)
std = df.groupby('continent').std()
display("그룹에 대한 표준편차 =", std)
```

(15) 국가별, 년도에 대한 GDP 테이블로 재구성

```python
df3 = pd.pivot_table(df, values='gdpPercap', index='year', columns='country')
display("df3=", df3)
```

(16) 년도별 대륙에 대한 인구수 평균 테이블로 재구성

```python
import numpy as np
```

```python
df4 = pd.pivot_table(df,values='pop', index='year', columns='continent', aggfunc=np.mean)
display("df4=", df4)
```

(17)년도별 대륙에 대한 GDP 평균 테이블로 재구성

```python
df5 = pd.pivot_table(df,values='gdpPercap', index='year', columns='continent', aggfunc=np.mean)
display("df5=", df5)
```

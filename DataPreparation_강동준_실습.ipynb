**(2) 데이터 가공**

(1)초등학교 교원수(시도/시/군/구), 초등학교 학생수(시도/시/군/구) 두 개의 엑셀파일을 읽어 테이블에 저장

```python
import pandas as pd
df_tea = pd.read_excel('초등학교_교원수_시도_시_군_구__20220921100227.xlsx')
df_stu = pd.read_excel('초등학교_학생수_시도_시_군_구__20220921100829.xlsx')
```

```python
df_tea.head()
```

```python
df_stu.head()
```

(2) (1)에서 얻은 두 테이블로 2022년 학생수, 교원수 테이블 생성

```python
df_te_2022 = df_tea.set_index('행정구역별')
df_te_2022 = df_te_2022['2022']
display(df_te_2022)
```

```python
df_st_2022 = df_stu.set_index('행정구역별')
df_st_2022 = df_st_2022['2022']
display(df_st_2022)
```

```python
df_merge = pd.concat([df_st_2022,df_te_2022], axis=1)
```

```python
display(df_merge)
```

```python
df_merge.columns = ['학생수','교원수']
```

```python
display(df_merge)
```

(3)교원 1인당 학생수 속성 추가 (학생수 / 교원수)

```python
df_merge['교원 1인당 학생수'] =  df_merge['학생수']/ df_merge['교원수']
display(df_merge)
```

(4)교원1인당 학생수 역순으로 정렬

```python
df_merge = df_merge.sort_values(by='교원 1인당 학생수',ascending=[False])
display(df_merge)
```

(5)주요 3개 도시 서울,부산,대구 데이터만 추출

```python
df_merge = df_merge.loc[['서울특별시','부산광역시','대구광역시']]
display(df_merge)
```

 (6)전체 년도에 대한 교원1인당 학생수 계산

```python
df_te_all = df_tea.set_index('행정구역별')
df_st_all = df_stu.set_index('행정구역별')
```

```python
df = df_st_all / df_te_all
display(df)
```

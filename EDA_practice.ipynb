# 1. Anscombe's quartet 데이터 셋 분석

Anscombe's quartet 데이터 셋을 이용 

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set style of plotting area
import matplotlib.style as style
style.use('seaborn')
```

```python
import io
import pandas as pd

anscombe = pd.read_csv(io.StringIO(uploaded['anscombe.csv'].decode('utf-8')))
```

```python
anscombe.columns = anscombe.columns.str.split('_', expand=True)
anscombe.columns = anscombe.columns.swaplevel(1,0)

anscombe
```

각 데이터 셋에 대한 요약통계 확인

```python
anscombe["I"].describe()
```

```python
anscombe["II"].describe()
```

```python
anscombe["III"].describe()
```

```python
anscombe["IV"].describe()
```

각 데이터 셋에 대한 상관계수 확인

```python
print("Correlation coefficient for:\n")

print("Dataset I: " + str(anscombe["I"]["x"].corr(anscombe["I"]["y"])))
print("Dataset II: " + str(anscombe["II"]["x"].corr(anscombe["II"]["y"])))
print("Dataset III: " + str(anscombe["III"]["x"].corr(anscombe["III"]["y"])))
print("Dataset IV: " + str(anscombe["IV"]["x"].corr(anscombe["IV"]["y"])))
```

```python
import io
import pandas as pd

fitness = pd.read_csv(io.StringIO(uploaded['physicalFitness.csv'].decode('utf-8')))
```

```python
(2)남녀 데이터를 분리 
```

```python
woman.describe()
```

(4)전체 데이터의 키와 몸무게의 상관관계 확인

```python
print("Correlation coefficient for:\n")

print(str(fitness["height"].corr(fitness["weight"])))
```

(5)성별에 대한 연령별 평균 BMI 테이블 재구성하여 요약통계 확인

```python
table1 = pd.pivot_table(fitness, values='BMI', index=['sex'],
                    columns=['ageGroup'], aggfunc=np.mean)
```

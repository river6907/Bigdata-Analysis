## 데이터 시각화 (data visualization)

### (1)데이터 시각화(data visualization)

  데이터 분석 결과를 쉽게 이해할 수 있도록 시각적으로 표현하고 전달되는 과정  
  데이터 시각화의 목적은 그래프라는 수단을 통해 정보를 명확하고 효과적으로 전달하는 것 (Friedman, 2008).    

### (2)데이터 시각화의 중요성

  - Anscombe's quartet   

   기술통계량은 유사하나 분포가 다른 4개의 데이터 셋 (11개 x, y 값) 


### (3)그래프를 통한 데이터 분석 

 - Histogram, Boxplot : 단일 수치형 변수의 분포 확인  
 - Bar, Pie : 범주형 변수에 대한 비교  
 - Scatter : 수치형 변수들간의 관계 확인, outlier 확인  
 - Scatter matrix : 여러 변수들 사이의 모든 가능한 scatter를 한장에 표현하여 변수들 사이의 관계를 신속하게 확인가능

## simple plot

```python
import numpy as np
import matplotlib.pyplot as plt

#plot() 함수를 이용한 그래프 그리기 
x = np.linspace(0, 3, 100) #0 ~ 3사이에서 균일한 간격의 점 100개로 구성된 데이터
x1 = x
x2 = x**2
x3 = x**3
plt.plot(x, x1, label='linear', color='r')    #1차 함수 (x축=x, y축=x1), 빨간색
plt.plot(x, x2, label='quadratic', color='g') #2차함수 (x축=x, y축=x2), 녹색
plt.plot(x, x3, label='cubic', color='b')     #3차함수 (x축=x, y축=x3), 파란색
plt.xlabel('x label')  #x축 라벨
plt.ylabel('y label')  #y축 라벨
plt.title('Simple plot') #제목
plt.legend()  #범례
plt.show()  #플롯 그리기
```

### Bar plot
범주형에 따른 개수, 퍼센트 등을 이용하여  막대그래프로 시각화  


범주형 변수(categorical variable) : 국가, 성별, 혈액형 등  그룹을 분류하기 위한 비연속적인 값(discrete)을 가지는 변수  

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

#단일값 그래프
df1 = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
df1.plot.bar(x='lab', y='val', title="Bar plot",rot=0) #x : 축 레이블, y: 값 (범례), title:제목, rot : x축 출력값 기울기
plt.show()

#다중값 그래프
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
          'rabbit', 'giraffe', 'coyote', 'horse']
df2 = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
display(df2)
df2.plot.bar(rot=0)
plt.show()
#세부속성 y값 순서, x축 값 기울기 조정
df2.plot.bar(y=['lifespan', 'speed'], rot=45) 
plt.show()
#단일 y값 출력
df2.plot.bar(y=['speed'], rot=0) 
plt.show()
#subplot으로 분리된 그래프
axes = df2.plot.bar(rot=0, subplots=True,  figsize=(8, 4))   #서브플롯, 그림 크기 (폭8, 높이 4)
plt.show()
```

### Pie plot

범주형에 따른 개수, 퍼센트 등을 이용하여  원그래프로 시각화  


```python
df = pd.DataFrame(3 * np.random.rand(4, 2),
                  index=['a', 'b', 'c', 'd'], columns=['x', 'y'])
display(df)

# x 그래프
df.plot.pie(y='x', title = 'pie plot (x)')  #x만 그리기
# y 그래프
df.plot.pie(y='y', title = 'pie plot (y)',  autopct='%.2f%%')  #y만, % 삽입 그리기

#x, y그래프
df.plot.pie( title = 'pie subplot', subplots=True, figsize=(8, 4))  

#다중값 df2에 대한 그래프
df2.plot.pie( title = 'pie plot (speed-lifespan)', subplots=True, figsize=(14, 8))

plt.show()
```

### histogram
특정 구간에 속하는 데이터의 빈도수(frequency)를 막대그래프로 표현  
수치 데이터의 분포를 표현
- 히스토그램 분석시 확인사항
  - 전반적인 분포모양 확인 
    :정규분포형태(종모양), 치우친 모양, 피크 개수 등 확인
  - 분포를 벗어나는 데이터(outlier)는 없는지 확인
  - 분포가 치우친 경우 정규분포로 변환 
  - 빈도구간(bin)을 조정하여 자세하거나 거친 부분 조정

```python
df3 = pd.DataFrame({
     'x': [1, 2, 2.5, 3, 3.5, 4, 5],
     'y': [4, 4, 4.5, 5, 5.5, 6, 6],
 })

#histogram
df3.plot.hist(y='x')
df3.plot.hist(y='x', bins=3)

#다중 histogram 
df3.hist()
df3.hist(bins=3)

plt.show()
```

### density  
  확률밀도함수 Probability Density  Function, pdf)  
  커널밀도추정( kernel density estimation,KDE) 

```python
df3 = pd.DataFrame({
     'x': [1, 2, 2.5, 3, 3.5, 4, 5],
     'y': [4, 4, 4.5, 5, 5.5, 6, 6],
 })

df3.hist(bins=3)

#Generate Kernel Density Estimate plot using Gaussian kernels.
df3.plot.density()
df3.plot.kde()
plt.show()
```

### boxplot
   사 분위수(quartiles)를 통해 수치 데이터 분포를 확인하는 그래프 

```python
df = pd.DataFrame(np.random.randn(10,4), columns=['A','B', 'C', 'D'])

display(df)

df.boxplot()
df.plot.density()
plt.show()
```

### 산점도(Scatter plot)
두 수량형 변수의 값을 x, y 좌표로 하는 점으로 표현  
X값을 입력변수, y값을 출력변수로 관계 확인  
두 변수의 선형적 상관관계 분석에 활용  

- 산점도(Scatter plot) 분석 시  확인사항
 - 데이터 개수가 클 경우 1000개 정도로 샘플링 하여 표시
 - x, y 값의 분포가 넓을 경우 로그변환 필요
 - x, y변수의 입력, 출력관계가 반영되었는지 확인
 - 두 변수 사이의 상관관계가 강한지 확인
 - 두 변수 사이의 선형 상관관계가 있는지 확인 

```python
# 속성 2개 비교
df2 = pd.DataFrame([[7.1, 3.4, 0], [6.7, 4.0, 0], [5.6, 9.3, 1],
                  [5.3, 7.8, 1], [5.8, 4.6, 2],[2.5, 3.2, 2]],
                  columns=['length', 'width', 'types'])
display(df2)
df2.plot.scatter(x='length',  y='width',   c='Blue')  #c로 색상지정
df2.plot.scatter(x='length',  y='width',  c='types', colormap='plasma') #type을 칼라맵의 인덱스로 지정
df2.plot.scatter(x='length',  y='width',   marker='^', s= 60 ) #마커모양과 크기 지정
df2.plot.scatter(x='length',  y='width', s= df2['types'] * 50) #type의 값에따라 크기변경

# 속성 4개 비교
df3 = pd.DataFrame(np.random.rand(20, 4) * 5, columns=['a', 'b', 'c', 'd'])
display(df3)
ax = df3.plot.scatter(x='a', y='b', color='Skyblue', label='Group 1');
ax = df3.plot.scatter(x='c', y='d', color='Orange', label='Group 2', ax=ax); #ax plot을 결합하여 그리기

plt.show()
```

### Scatter Matrix
여러 변수에 대한 scatter plot의 matrix

```python
df4 = pd.DataFrame(np.random.rand(20, 4) * 5, columns=['a', 'b', 'c', 'd'])

df4['class'] =np.round(np.random.rand(20,1) * 5)  #class 속성 생성
df4['class']= df4['class'].astype("category")     #범주형으로 타입변경

display(df4)

pd.plotting.scatter_matrix(df4, s=60, diagonal='kde', c= df4['class'], cmap='spring') #class로 spring색상 설정
pd.plotting.scatter_matrix(df4, s=60, diagonal='hist', c= df4['class'], cmap= 'viridis')
plt.show()
```
